### Example using just matplotlib

```python
sample_x = [1,2,3,4,5]
sample_y = [34,65,23,78,43]

# all figures should start with the next two lines
fig = plt.figure()    #creates an empty figure
ax = fig.add_subplot(111)  #creates axes on the figure. 111 = 1 vertical axis, 1 horiz axis, following should edit first plot
ax1 = ax.plot(sample_x,sample_y,label="my data") #ax.plot(xdata,ydata,optional properties....)


```

```python


#personalize the plot
ax.set_title('hi')
ax.set_ylabel('y-axis')
ax.set_xlabel('x-axis')
ax.grid()
ax.legend()

#in Jupyter notebook...it will keep updating the plot that printed first, until you make a new fig
```
